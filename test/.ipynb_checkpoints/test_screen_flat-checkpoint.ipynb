```python
import numpy
import xarray
```

```python
dem_filename = '/disk1/Data/MERIT_DEM/s20e120_dem.tif'
```

```python
screen_range = 0.03
roi_lat = -17.95
roi_lon = 123.45
```

```python
dem_s = xarray.open_rasterio(dem_filename)
```

```python
dem_s
```

```python
resolution = dem_s.transform[0]
roi_lons = numpy.arange(roi_lon-screen_range, roi_lon+screen_range, resolution)
roi_lats = numpy.arange(roi_lat+screen_range, roi_lat-screen_range, -resolution)
```

```python
roi_dem_s = dem_s.interp(x=roi_lons, y=roi_lats, method='nearest')
```

```python
roi_dem_v = numpy.array(roi_dem_s)[0]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(roi_dem_v)
```

## calculate max-min of elevation

```python
roi_dem_1d = roi_dem_v.flatten()
roi_dem_1d = numpy.delete(roi_dem_1d, numpy.where(roi_dem_1d<50))
```

```python
def is_flat(dem_array):
    if dem_array.max() - dem_array.min() > 100:
        return 0
    else:
        return 1
```

```python
is_flat(roi_dem_1d)
```

## calculate slope

```python
from xrspatial import slope
```

```python
roi_slope = slope(xarray.DataArray(roi_dem_v))
```

```python
plt.imshow(roi_slope)
plt.colorbar()
```

```python
def is_flat_slope(dem_array):
    dem_slope = slope(xarray.DataArray(roi_dem_v))
    flat_num = numpy.sum(dem_slope<=30)
    noflat_num = numpy.sum(dem_slope>30)
    if flat_num/noflat_num > 2.:
        return 1
    else:
        return 0
```

```python
is_flat_slope(roi_dem_1d)
```
