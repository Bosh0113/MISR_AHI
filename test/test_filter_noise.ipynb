##  slope identifer

```python
def identifer(data):
    down, up = numpy.nanpercentile(data, [0, 75])
    IQR = up-down
    lower_limit = down - 1.5*IQR
    upper_limit = up + 1.5*IQR
    result = numpy.where(data > upper_limit, numpy.nan, data)
    result = numpy.where(result < lower_limit, numpy.nan, result)
    return result
```

```python
slope_array = numpy.array(Y)/numpy.array(X)
slope_array_filtered = numpy.array(identifer(slope_array))
```

```python
show_x = (slope_array_filtered*0+1)*X
show_y = (slope_array_filtered*0+1)*Y
show_x = show_x[~numpy.isnan(show_x)]
show_y = show_y[~numpy.isnan(show_y)]
```

```python
numpy.polyfit(show_x, show_y, deg=1)
```

```python
pearsonr(show_x, show_y)
```

```python
plt.scatter(show_x, show_y)
```

## Upper quartile

```python
def identifer_up4_iqr(data):
    down, up = numpy.nanpercentile(data, [0, 75])
    IQR = up-down
    lower_limit = down - 1.5*IQR
    upper_limit = up + 1.5*IQR
    result = numpy.where(data > upper_limit, numpy.nan, data)
    result = numpy.where(result < lower_limit, numpy.nan, result)
    return result
```

```python
up4_x = identifer_up4_iqr(X)
up4_y = identifer_up4_iqr(Y)
```

```python
plt.scatter(up4_x, up4_y)
```

```python
v_mask = 0*(up4_x*up4_y)+1
v_mask = v_mask.flatten()
x_1d = up4_x.flatten()
y_1d = up4_y.flatten()
x_1d = x_1d[~numpy.isnan(v_mask)]
y_1d = y_1d[~numpy.isnan(v_mask)]
numpy.polyfit(x_1d, y_1d, deg=1)
```

```python
pearsonr(x_1d, y_1d)
```

```python
def detect_outliers(data):
    mean = numpy.mean(data)
    std = numpy.std(data)
    threshold = 3 * std
    outliers = []
    for x in data:
        if abs(x - mean) > threshold:
            outliers.append(numpy.nan)
        else:
            outliers.append(x)
    return outliers
```

```python
diff_array = abs(numpy.array(Y)-numpy.array(X))/abs(numpy.minimum(numpy.array(X), numpy.array(Y)))
diff_array_filtered = numpy.array(detect_outliers(diff_array))
```

```python
show_x = (diff_array_filtered*0+1)*X
show_y = (diff_array_filtered*0+1)*Y
show_x = show_x[~numpy.isnan(show_x)]
show_y = show_y[~numpy.isnan(show_y)]
```
