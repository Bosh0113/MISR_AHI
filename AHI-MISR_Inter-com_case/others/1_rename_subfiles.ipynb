```python
import os
import numpy
import re
```

```python
WORK_SPACE = os.getcwd()
```

```python
folder_l1_list = ['0', '26', '45', '60', '70']
folder_l2_list = ['0', '1']

for folder_l1 in folder_l1_list:
    folder_l1_path = os.path.join(WORK_SPACE, folder_l1)
    for folder_l2 in folder_l2_list:
        folder_l2_path = os.path.join(folder_l1_path, folder_l2)
        roi_folder_list = os.listdir(folder_l2_path)
        for roi_folder in roi_folder_list:
            roi_folder_path = os.path.join(folder_l2_path, roi_folder)
            roi_folder_list = os.listdir(roi_folder_path)
            for roi_file in roi_folder_list:
                roi_filename = os.path.join(roi_folder_path, roi_file)
                str_re = r'(\S+)_4AC_record.npy'
                ac_matchObj = re.search(str_re, roi_file)
                if ac_matchObj:
                    n_roi_filename = os.path.join(roi_folder_path, roi_folder + '_4AC_record.npy')
                    os.rename(roi_filename, n_roi_filename)
                    print(roi_filename, n_roi_filename)
                
                str_re = r'(\S+)_4AC_record.txt'
                ac_matchObj = re.search(str_re, roi_file)
                if ac_matchObj:
                    n_roi_filename = os.path.join(roi_folder_path, roi_folder + '_4AC_record.txt')
                    os.rename(roi_filename, n_roi_filename)
                    print(roi_filename, n_roi_filename)
                    
                str_re = r'(\S+)_matched_record.npy'
                ac_matchObj = re.search(str_re, roi_file)
                if ac_matchObj:
                    n_roi_filename = os.path.join(roi_folder_path, roi_folder + '_matched_record.npy')
                    os.rename(roi_filename, n_roi_filename)
                    print(roi_filename, n_roi_filename)
                
                str_re = r'(\S+)_MISR_AHI_RAY_MATCH_RECORD.txt'
                ac_matchObj = re.search(str_re, roi_file)
                if ac_matchObj:
                    n_roi_filename = os.path.join(roi_folder_path, roi_folder + '_MISR_AHI_RAY_MATCH_RECORD.txt')
                    os.rename(roi_filename, n_roi_filename)
                    print(roi_filename, n_roi_filename)
```
